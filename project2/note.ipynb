```python
with open ('employees.csv',"w") as f:
    writer = csv.DictWriter(f, fieldnames=["id", 'name', 'email', "gender", "active"])
    writer.writeheader()
    for i in employees['data']:
        id = i['id']
        name = i['name']
        email = i['email']
        gender = i['gender']
        active = True if i['status'] == "active" else False
        writer.writerow(
            {
                "id" : id,
                "name" : name,
                "email" : email,
                "gender" : gender,
                "active" : active
            }
        )
```
